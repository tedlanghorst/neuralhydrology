```python
# from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import xarray as xr

data_path = "./configs/selected_basins/runs/test_run_2010_131408/test/model_epoch050/test_results.p"

# Load the Xarray dataset
dataset = pd.read_pickle(data_path)

print(f"loaded {len(dataset)} basins")
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(12,12))
axes = axes.ravel()

for i, (key,value) in enumerate(dataset.items()):
    date = value['1D']['xr'].variables['date'].values
    obs = value['1D']['xr'].variables['turbidity (NTU)_obs'].values
    sim = value['1D']['xr'].variables['turbidity (NTU)_sim'].values
    
    axes[i].plot(date,sim, zorder=1)
    axes[i].scatter(date,obs, color='red', zorder=2)
    axes[i].set_title(key)
    
plt.tight_layout()
plt.show()
```
